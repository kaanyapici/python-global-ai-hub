```python
import pandas as pd
df = pd.DataFrame(columns=["Ad-Soyad", "Okul no", "Sınav Puanı", "Harf Notu"])
parts = int(input("Sınıf mevcudunu girin: "))

for _ in range(parts):
    oad = input("Öğrencinin Adı-Soyadı: ")
    ono = input("Öğrencinin Okul Numarası: ")
    Mat = int(input("Öğrencinin sınav puanı: "))
    if Mat>80: hn = "A"
    elif Mat>60: hn = "B"
    elif Mat>40: hn = "C"
    else: hn = "D"
    df1 = pd.DataFrame(data=[[oad,ono,Mat,hn]],columns=["Ad-Soyad", "Okul no", "Sınav Puanı", "Harf Notu"])
    df = pd.concat([df,df1], axis=0)
df.index = range(len(df.index))
```
